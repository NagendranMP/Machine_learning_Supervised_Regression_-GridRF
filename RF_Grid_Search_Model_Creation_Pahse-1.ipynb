```python
import pandas as pd
```

```python
df=pd.read_csv("50_Startups.csv")
df
```

```python
df=pd.get_dummies(df,drop_first=True).astype(int)
df
```

```python
Independent=df.drop("Profit",axis=1)
Dependent=df["Profit"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Independent,Dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
import joblib
joblib.dump(sc,"scalar.pkl")
```

```python
from sklearn.ensemble import RandomForestRegressor
RF=RandomForestRegressor()
```

```python
para={"criterion":["squard_error","absolute_error","friedman_mse","poisson"],
     "n_estimators":[100],
     "max_features":["sqrt","log2"]}
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(RF,para,refit=True,verbose=3,n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
re=grid.cv_results_
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score
R2_score=r2_score(y_test,y_pred)
print(R2_score)
```

```python
df=pd.DataFrame.from_dict(re)
df
```

```python
import pickle
filename="Finalized_Model.sav"
pickle.dump(grid,open(filename,"wb"))
```
